```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import Table, Column, MetaData, Integer, String, Date, Float, BigInteger

db_engine = create_engine("postgresql+psycopg2://postgres:admin1234@localhost:5432/bootcamp_2508")

metadata = MetaData()

person_schema = Table(
  "stock_symbol",
  metadata,
  Column("Symbol", String(20), primary_key=True),
  Column("Security", String(200), nullable=False),
  Column("GICS Sector", String(200), nullable=False),
  Column("GICS Sub-Industry", String(200), nullable=False),
  Column("Headquarters Location", String(200), nullable=False),
  Column("Date added", Date, nullable=False),
  Column("CIK", BigInteger, nullable=False),
  Column("Founded", String(100), nullable=False)
)

metadata.create_all(db_engine)

df = pd.read_csv("stockSymbol.csv")

# df (dob, from str to date)
df['Date added'] = pd.to_datetime(df['Date added'].astype(str), format="%Y-%m-%d").dt.date
# write df to db (check dataframe data type same as database data type)
df.to_sql('stock_symbol', db_engine, index=False, if_exists='replace')

df_symbol_only = pd.read_sql('select "Symbol" from stock_symbol', con=db_engine)
df_symbol_only.columns = df_symbol_only.columns.str.lower()

df_symbol_only['symbol'] = df_symbol_only['symbol'].str.replace('.', '-')

df_symbol_only.to_sql('stock_symbol_list', db_engine, index=False, if_exists='append')
```
