```python
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({
    'text_column': ['hello world', 'python programming', 'data analysis'],
    'numeric_column': [1, 2, 3]
})

print("Original DataFrame:")
print(df)

# Replace 'o' with 'X' in 'text_column'
df['text_column'] = df['text_column'].str.replace('o', 'X')

print("\nDataFrame after replacing 'o' with 'X':")
print(df)

# Replace multiple characters using regex
df['text_column'] = df['text_column'].str.replace(r'[aeiou]', '*', regex=True)

print("\nDataFrame after replacing vowels with '*':")
print(df)
```
